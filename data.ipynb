```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importación de los datos
dataset = "./Combustibles_Precios_Argentina_2016-2023 - Copy.csv"
```

```python
# Lectura y visualización de los datos
data = pd.read_csv(dataset)
data
```

```python
data.info()
```

```python
# Se elimina la columna "Año y Mes"
data = data.drop(columns=['Año y Mes'])

```

```python
# Verificación de las columnas
data.columns
```

```python
# Asignamos nuevos nombres para las columnas
nombre_columna = {
    'Empresa ID': 'Id_empresa',
    'Indentificacion Fiscal': 'Id_fiscal',
    'Producto ID': 'Id_producto',
    'ID Horario': 'Id_horario',    
    'Producto ': 'Producto',
    'Horario Tipo': 'Tipo_horario',
    'ID Empresa Bandera': 'Id_empresa_bandera',
    'Empresa Bandera': 'Empresa_bandera',
    'Longitud;': 'Longitud'
}
data = data.rename(columns=nombre_columna)
```

```python
data
```

```python
# Elimino caracter de la columna Longitud
data['Longitud'] = data['Longitud'].str.replace(';', '')

data
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
# Elimino las columnas complementamente nulas
data = data.dropna(subset=['Id_empresa', 'Id_fiscal', 'Empresa', 'Direccion', 'Localidad', 
                           'Provincia', 'Region', 'Id_producto', 'Producto', 'Id_horario', 
                           'Tipo_horario', 'Precio', 'Fecha', 'Id_empresa_bandera', 
                           'Empresa_bandera', 'Latitud', 'Longitud'])

data

```

```python
# Verificación de valores nulos
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
# Asignación  de variables y tipos
data['Longitud'] = data['Longitud'].astype(float)
data['Id_empresa'] = data['Id_empresa'].astype(int)
data['Id_producto'] = data['Id_producto'].astype(int)
data['Id_horario'] = data['Id_horario'].astype(int)
data['Id_empresa_bandera'] = data['Id_empresa_bandera'].astype(int)

# Convertir la columna de fecha a tipo datetime
data['Fecha'] = pd.to_datetime(data['Fecha'], format='%d/%m/%Y')
```

```python
data
```

```python
data.describe()
```

```python
# Analisis de precios por provincia

precio_por_provincia = round(data['Precio'].groupby(data['Provincia']).mean().reset_index(), 2)
precio_por_provincia
```

```python
precio_provincia = precio_por_provincia.sort_values(by='Precio', ascending=True)

plt.figure(figsize=(14, 6))
bars = plt.bar(precio_provincia['Provincia'], precio_provincia['Precio'], color='lightblue')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 2),
             ha='center', va='bottom')
plt.xlabel('Provincia')
plt.ylabel('Precio promedio')
plt.title('Precio promedio por provincia (orden ascendente)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Agrupación de empresas según empresa bandera para conocer la cantidad
cant_emp_bandera = data['Empresa'].groupby(data['Empresa_bandera']).count().reset_index()
cant_emp_bandera
```

```python
# Precio promedio por empresa
precio_por_empresa = round(data['Precio'].groupby(data['Empresa_bandera']).mean().reset_index(), 2)
precio_por_empresa
```

```python
# Representación gráfica de precio promedio por empresa
precio_medio_por_empresa = data.groupby('Empresa_bandera')['Precio'].mean().sort_values()

plt.figure(figsize=(10, 6))
precio_medio_por_empresa.plot(kind='bar')
plt.title('Precio Medio por Empresa')
plt.xlabel('Empresa')
plt.ylabel('Precio Medio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Valores máximos alcanzados según últimos registros
precio_maximo_por_producto = data.groupby('Producto')['Precio'].max().reset_index()
precio_maximo_por_producto
```

```python
# Valores mínimos según comienzo de los registros
precio_minimo_por_producto = data.groupby('Producto')['Precio'].min().reset_index()
precio_minimo_por_producto
```

```python
# Promedio del precio por mes
precios_por_mes = data.groupby(data['Fecha'].dt.to_period('M'))['Precio'].mean()
plt.figure(figsize=(10, 6))
precios_por_mes.plot(marker='o', linestyle='-')
plt.title('Evolución del Precio Promedio Mensual')
plt.xlabel('Fecha (Mes)')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Precio promedio por año 
precios_por_año = data.groupby(data['Fecha'].dt.to_period('Y'))['Precio'].mean()

precios_por_año_df = precios_por_año.reset_index()

precios_por_año_df['Fecha'] = precios_por_año_df['Fecha'].astype(str)

plt.figure(figsize=(10, 6))
plt.plot(precios_por_año_df['Fecha'], precios_por_año_df['Precio'], marker='o', linestyle='-')
plt.title('Precio Promedio por Año')
plt.xlabel('Año')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
# Agrupamos por región y calculamos el precio promedio
bars = precios_por_region = data.groupby('Region')['Precio'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(precios_por_region['Region'], precios_por_region['Precio'])
plt.title('Precio Promedio por Región')
plt.xlabel('Región')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)

plt.show()
```

```python
# Media del precio por producto
promedio_por_producto = data['Precio'].groupby(data['Producto']).mean().reset_index()
promedio_por_producto
```

```python

precios_por_producto = data.groupby('Producto')['Precio'].mean().reset_index()

plt.figure(figsize=(10, 6))
bars = plt.bar(precios_por_producto['Producto'], precios_por_producto['Precio'])
plt.title('Precio Promedio por Producto')
plt.xlabel('Producto')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)

for bar, precio in zip(bars, precios_por_producto['Precio']):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 2, round(precio, 2), ha='center', color='black', fontsize=9)

plt.show()
```

```python
# Variación del precio del combustible a los largo d elos años
precio_producto_anio = data.groupby(['Producto', data['Fecha'].dt.year])['Precio'].mean().reset_index()

plt.figure(figsize=(10, 6))
for producto in precio_producto_anio['Producto'].unique():
    plt.plot(precio_producto_anio[precio_producto_anio['Producto'] == producto]['Fecha'],
             precio_producto_anio[precio_producto_anio['Producto'] == producto]['Precio'],
             label=producto)

plt.title('Variación del precio por producto a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Precio promedio')
plt.legend()
plt.grid(True)
plt.xticks(precio_producto_anio['Fecha'].unique()) 
plt.show()
```
