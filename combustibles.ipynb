```python
# Importamos las librearías con las cuales vamos a trabajar y manipular el dataset
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Se importa el dataset.
dataset = "./Historico-precios-2001-2024-Argentina.csv"
```

```python
# Lectura y visualización de la data.
data = pd.read_csv(dataset)
data
```

```python
data.info()
```

```python
columna_nombre = {
    'fecha_vigencia_regsitro': 'fecha_vigencia_registro'
}

data = data.rename(columns=columna_nombre) 
data
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
# Eliminar columnas nulas
data_nula = data.drop(columns=['horario', 'fecha_vigencia_registro'])
data = data_nula
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
# Precio promedio por Provincia.

precio_provincia = data['precio'].groupby(data['provincia']).mean().reset_index()
precio_provincia
```

```python
# Representación gráfica de los precios por provincia.
precio_por_provincia = precio_provincia.sort_values(by='precio', ascending=False)


plt.figure(figsize=(14, 6))
bars = plt.bar(precio_por_provincia['provincia'], precio_por_provincia['precio'], color='lightblue')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 2),
             ha='center', va='bottom')
plt.xlabel('Provincia')
plt.ylabel('Precio promedio')
plt.title('Precio promedio por provincia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Precio promedio por empresa
precio_empresa = data['precio'].groupby(data['nombre_empresa']).mean().reset_index()
precio_empresa
```

```python
precio_por_empresa = precio_empresa.sort_values(by='precio', ascending=False).head(10)


plt.figure(figsize=(14, 6))
bars = plt.bar(precio_por_empresa['nombre_empresa'], precio_por_empresa['precio'], color='lightblue')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 2),
             ha='center', va='bottom')
plt.xlabel('Empresa')
plt.ylabel('Precio promedio')
plt.title('Precio promedio por Empresa')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
